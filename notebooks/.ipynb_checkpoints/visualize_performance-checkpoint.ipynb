```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'axes.labelsize': 'large'})
```

### Load Data

```python
path = '../model_data/'
filepath = path + 'best_maestro_model_2_1_512_0pt4.csv'
df_base = pd.read_csv(filepath, index_col = 'Epochs')
filepath = path + 'best_maestro_model_2_1_512_0pt4_lr_5e-04.csv'
df_lr = pd.read_csv(filepath, index_col = 'Epochs')
filepath = path + 'best_maestro_model_2_1_512_0pt4_cn_1pt0.csv'
df_cn_1 = pd.read_csv(filepath, index_col = 'Epochs')
filepath = path + 'best_maestro_model_2_1_512_0pt4_cn_0pt5.csv'
df_cn_pt5 = pd.read_csv(filepath, index_col = 'Epochs')
filepath = path + 'best_maestro_model_2_1_512_0pt4_lr_5e-04_cn_1pt0.csv'
df_lr_cn_1 = pd.read_csv(filepath, index_col = 'Epochs')
filepath = path + 'best_maestro_model_2_1_512_0pt3_lr_5e-04_cn_1pt0.csv'
df_lr_cn_1_do_pt3 = pd.read_csv(filepath, index_col = 'Epochs')
filepath = path + 'best_maestro_model_2_1_512_0pt5_lr_5e-04_cn_1pt0.csv'
df_lr_cn_1_do_pt5 = pd.read_csv(filepath, index_col = 'Epochs')
```

```python
var_name_to_axis_name = {'maestro_loss': 'Training Loss', 'f1_score': 'Training F-measure', \
                         'precision': 'Training Precision', 'recall': 'Training Recall', \
                         'dur_error': 'Training Duration Relative Error', 'dur_loss': 'Training Duration Loss', \
                         'val_maestro_loss': 'Validation Loss', 'val_f1_score': 'Validation F-measure', \
                         'val_precision': 'Validation Precision', 'val_recall': 'Validation Recall', \
                         'val_dur_error': 'ValidationDuration Relative Error', \
                         'val_dur_loss': 'Validation Duration Loss'}
```

```python
def plot_comparison(dfs, labels, var = 'val_maestro_loss', title = None, ylim = None):
    """plot lines (maximum 6) for var versus Epoch, one for each dataframe in a list of 
    dataframes(dfs) with corresponding labels. Add title (str) and ylim if passed."""
    if (len(labels) != len(dfs)):
        print('dfs and labels are different length, abort.')
        return
    colors = ['r', 'b', 'k', 'c', 'y', 'm']
    if (len(dfs) > len(colors)):
        print('Input more dataframes than available colors, abort.')
    plt.figure(figsize = (12, 6))
    for i in range(len(dfs)):
        plt.plot(dfs[i][var], colors[i], label = labels[i])
    plt.xlabel('Epoch')
    plt.ylabel(var_name_to_axis_name[var])
    if (type(title) == str):
        plt.title(title)
    if (type(ylim) == tuple):
        plt.ylim(ylim)
    plt.legend()
```

```python
plot_comparison([df_base, df_lr, df_cn_1, df_cn_pt5], labels = ['base', 'lr = 5e-4', 'clipnorm = 1.0', 'clipnorm = 0.5'], \
                var = 'maestro_loss', title = 'Figure 1')
```

In Figure 1, we plot the Training Loss across Epochs. Surprisingly, the base model doesn't learn quite as fast as the models with clipnorm set and the lower clipnorm value (0.5) failed earlier than the higher value. The second surprise may have something to do with the blip near Epochs 7-10. Unsurprisingly, the lower learning rate (half default) learns the slowest. Interestingly, it learns the training data better than the other models.

```python
plot_comparison([df_base, df_lr, df_cn_1, df_cn_pt5], labels = ['base', 'lr = 5e-4', 'clipnorm = 1.0', 'clipnorm = 0.5'], \
                var = 'val_maestro_loss', title = 'Figure 2')
```

In Figure 2, we plot the Validation Loss versus time for the same models. The same trends can be seen, but the training is much noisier since the network is not minimizing Validation Loss. Also note that while the lr = 5.e-4 model learned the training set better, the clipnorm = 1.0 model actually performed better on the validation set in fewer epochs.

```python
plot_comparison([df_lr, df_cn_1, df_lr_cn_1], labels = ['lr = 5e-4', 'clipnorm = 1.0', 'both'], \
                var = 'maestro_loss', title = 'Figure 3')
```

Surprisingly, using both clipnorm = 1.0 and lr = 5e-4 causes an earlier failure and consequently does not perform as well on the training loss.

```python
plot_comparison([df_lr, df_cn_1, df_lr_cn_1], labels = ['lr = 5e-4', 'clipnorm = 1.0', 'both'], \
                var = 'val_maestro_loss', title = 'Figure 4')
```

Using 'both', however, does yield the lowest validation loss. Locking-in these settings, we next look at the effect of dropout_rate.

```python
plot_comparison([df_lr_cn_1_do_pt3, df_lr_cn_1, df_lr_cn_1_do_pt5], labels = ['dropout_rate = 0.3', 'dropout_rate = 0.4', \
                'dropout_rate = 0.5'], var = 'maestro_loss', title = 'Figure 5')
```

As expected, the training loss is more quickly learned for lower dropout_rate and it also delays the NaN failure. I cannot at this time explain the local increases in training loss near Epoch 7-12 but note that it also happened in Figure 1 with clipnorm = 0.5 during a similar stage in training.

```python
plot_comparison([df_lr_cn_1_do_pt3, df_lr_cn_1, df_lr_cn_1_do_pt5], labels = ['dropout_rate = 0.3', 'dropout_rate = 0.4', \
                'dropout_rate = 0.5'], var = 'val_maestro_loss', title = 'Figure 6')
```

While lowering the dropout_rate delayed NaN failure, it also increased overfitting which prevented it from outperforming the original dropout_rate (0.4) model on the validation set. Going the other direction (dropout_rate  0.5) also did not outperform because of its premature failure by NaN loss. I halt the hyperparameter tuning here. Our best model is that with lr = 5e-4, clipnorm = 1.0, and dropout_rate = 0.4.

```python
plt.figure(figsize = (12, 6))
plt.plot(df_lr_cn_1.index, df_lr_cn_1['maestro_loss'], 'r', label = 'Training total')
plt.plot(df_lr_cn_1.index, df_lr_cn_1['dur_loss'], 'b', label = 'Training duration component')
plt.plot(df_lr_cn_1.index, df_lr_cn_1['val_maestro_loss'], 'r--', label = 'Validation total')
plt.plot(df_lr_cn_1.index, df_lr_cn_1['val_dur_loss'], 'b--', label = 'Validation duration component')
plt.title('Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend()
```

Learning is very quick at first, levels out in the duration component for both sets (training and validation) before leveling out in the validation total loss near the end (NaN failure) where the total training loss was still steadily decreasing.

```python
plt.figure(figsize = (12, 6))
plt.plot(df_lr_cn_1.index, df_lr_cn_1['dur_error'], 'r', label = 'Training')
plt.plot(df_lr_cn_1.index, df_lr_cn_1['val_dur_error'], 'b', label = 'Validation')
plt.title('Duration Error')
plt.ylabel('Relative Error')
plt.xlabel('Epochs')
plt.legend()
```

The average difference between these two lines is quite remarkable and fortuitous. We reach a validation relative error in duration of just under 25%.

```python
plt.figure(figsize = (12, 6))
plt.plot(df_lr_cn_1.index, df_lr_cn_1['dur_error'], 'r', label = 'Training')
plt.plot(df_lr_cn_1.index, df_lr_cn_1['val_dur_error'], 'b', label = 'Validation')
plt.title('Duration Error')
plt.ylabel('Relative Error')
plt.xlabel('Epochs')
plt.legend()
```
